```python
import pandas as pd
from kozak_pipe import run_kozak

# You prepare df however you like (from CSV, DB, whatever)
df = pd.read_csv("data/Stock level signals.csv")

results = run_kozak(df)

def plot_results(results):
    df_ols_sr2   = results["df_ols_sr2"]
    df_ols_impr  = results["df_ols_impr"]
    df_uni_sr2   = results["df_uni_sr2"]
    df_uni_impr  = results["df_uni_impr"]

    # Then you can recreate your notebook plots:
    import matplotlib.pyplot as plt

    df_ols_sr2.plot(marker="o")
    plt.xlabel("Number of factors K")
    plt.ylabel("Average total SR$^2$ (annualized) - OLS")
    plt.show()

    df_uni_sr2.plot(marker="o")
    plt.xlabel("Number of factors K")
    plt.ylabel("Average total SR$^2$ (annualized) - UNI")
    plt.show()
```

```python

plot_results(results)
```

```python
recession_df = pd.read_csv("data/recession_stock_signals.csv")
crisis_df    = pd.read_csv("data/crisis_stock_signals.csv")
expansion_df = pd.read_csv("data/expansion_stock_signals.csv")
```

```python
rescession_results = run_kozak(recession_df)
plot_results(rescession_results)
```

```python
crisis_results    = run_kozak(crisis_df)
plot_results(crisis_results)
```
